```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import string
from google.colab import files

# Function to preprocess text
def preprocess_text(text):
    text = str(text)  # Convert to string to handle NaN values
    text = text.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    text = text.lower()  # Convert to lowercase
    return text

# Function to perform LDA topic modeling
def perform_lda(dataset, n_topics):
    try:
        # Create a CountVectorizer to convert text into a matrix of token counts
        vectorizer = CountVectorizer(stop_words='english')
        X = vectorizer.fit_transform(dataset)

        # Perform LDA
        lda_model = LatentDirichletAllocation(n_components=n_topics, random_state=42)
        lda_output = lda_model.fit_transform(X)

        return lda_output
    except ValueError as e:
        print("Error:", e)
        return None

# Upload dataset file
print("Upload the dataset file:")
uploaded = files.upload()

# Load the dataset with ISO-8859-1 encoding
dataset = pd.read_csv(next(iter(uploaded)), encoding='ISO-8859-1')

# Focus only on the specified columns: 'App', 'genre_id', 'description'
dataset = dataset[['App', 'genre_id', 'description']]

# Preprocess the description column
dataset['description'] = dataset['description'].apply(preprocess_text)

# List of all genres
genres = [
    (6000, 'Business'), (6001, 'Weather'), (6002, 'Utilities'), (6003, 'Travel'),
    (6004, 'Sports'), (6005, 'Social Networking'), (6006, 'Reference'), (6007, 'Productivity'),
    (6008, 'Photo & Video'), (6009, 'News'), (6010, 'Navigation'), (6011, 'Music'),
    (6012, 'Lifestyle'), (6013, 'Health & Fitness'), (6014, 'Games'), (6015, 'Finance'),
    (6016, 'Entertainment'), (6017, 'Education'), (6018, 'Books'), (6020, 'Medical'),
    (6021, 'Magazines & Newspapers'), (6022, 'Catalogs'), (6023, 'Food & Drink'), (6024, 'Shopping'),
    (6025, 'Stickers'), (7001, 'Action'), (7002, 'Adventure'), (7003, 'Arcade'), (7004, 'Board'),
    (7005, 'Card'), (7006, 'Casino'), (7007, 'Dice'), (7008, 'Educational'), (7009, 'Family'),
    (7010, 'Kids'), (7011, 'Music'), (7012, 'Puzzle'), (7013, 'Racing'), (7014, 'Role Playing'),
    (7015, 'Simulation'), (7016, 'Sports'), (7017, 'Strategy'), (7018, 'Trivia'), (7019, 'Word')
]

# Create an empty DataFrame to store all mappings
all_mappings = pd.DataFrame(columns=['App', 'genre_id', 'description', 'Topic (LDA with 5 topics)', 'Topic (LDA with 7 topics)'])

# Perform LDA for each genre
for genre_id, genre_name in genres:
    genre_data = dataset[dataset['genre_id'] == genre_id]

    # Perform LDA with 5 topics
    lda_output_5 = perform_lda(genre_data['description'], 5)

    if lda_output_5 is not None:
        # Perform LDA with 7 topics
        lda_output_7 = perform_lda(genre_data['description'], 7)

        # Add the topic mappings to the DataFrame
        genre_data['Topic (LDA with 5 topics)'] = np.argmax(lda_output_5, axis=1) + 1  # Adding 1 to start topics from 1
        genre_data['Topic (LDA with 7 topics)'] = np.argmax(lda_output_7, axis=1) + 1  # Adding 1 to start topics from 1

        # Append genre_data to all_mappings
        all_mappings = pd.concat([all_mappings, genre_data])

print("All mappings saved.")
# Save all_mappings to a CSV file
all_mappings.to_csv("all_genre_mappings.csv", index=False)
```
